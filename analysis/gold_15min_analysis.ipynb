```python
import pandas as pd

# Access CPI historical data (monthly)
data = '../data/gold_15min_data.csv'
df_data = pd.read_csv(data)
display(df_data.head())
print(df_data.columns)

# Check for anomalies
print("Gold Data:")
print(f"Number of rows: {len(df_data)}")
print(f"Start Date: {df_data['Date'].min()}")
print(f"End Date: {df_data['Date'].max()}")
print("\nMissing values per column:")
print(df_data.isnull().sum())
```

Display historical data (Gold futures 15-minute)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df_data = pd.read_csv('../data/gold_15min_data.csv')

# Convert Date to datetime
df_data['Date'] = pd.to_datetime(df_data['Date'])

# Plot
plt.figure(figsize=(14, 6))
plt.plot(df_data['Date'], df_data['Close'], color='gold', linewidth=1)
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.title('15-min Gold Price Over Time')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
